```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

import warnings

warnings.filterwarnings('ignore')
warnings.filterwarnings("ignore", message="numpy.ufunc size changed")
```

```python
data = pd.read_csv('files_for_lab/we_fn_use_c_marketing_customer_value_analysis.csv')
```

```python
# Get the numerical variables from our dataset.
```

```python
numericals = data.select_dtypes(np.number)
categoricals = data.select_dtypes(np.object)
```

```python
# Check using a distribution plot if the variables fit the theoretical normal 
# or exponential distribution.

for i in numericals.columns:
    
    sns.distplot(numericals[i])
    plt.show()
```

```python
# Check if any of the transformations (log-transform, etc.) 
# we have seen up to this point changes the result.



def box_cox_plot(data):
    
        
    from scipy.stats import boxcox
    import matplotlib.pyplot as plt
    
    for i in data:
        data[i] = data[i].map(lambda x: x+1)
        print(i)
        xt, lmbd = boxcox(data[i])
        sns.distplot(xt)
        plt.show()
    
    return 
```

```python
box_cox_plot(numericals)
```

```python
def log_plot(data):
    
    
    for i in data:
        data[i] = data[i].map(lambda x: x+1)
        x = np.log(data[i])
        sns.distplot(x)
        plt.show()
    
    return
```

```python
log_plot(numericals)
```
